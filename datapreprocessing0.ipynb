## outliers/extrema

## IQR method

```python
# prompt: use iqr method on car_data to find outliers

# Calculate Q1, Q3, and IQR for 'mileage'
Q1 = car_data['Mileage'].quantile(0.25)
Q3 = car_data['Mileage'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = car_data[(car_data['Mileage'] < lower_bound) | (car_data['Mileage'] > upper_bound)]

# Print or further process the outliers
print("Outliers in 'Mileage' column using IQR method:")
outliers
```

no outlier data in good state
